# 필요 라이브러리 로드

```python
import numpy as np
import pandas as pd

from tqdm import tqdm_notebook
from pickle5 import pickle
```

# 데이터 로드

```python
errorInfoPath = '/content/drive/MyDrive/Data enhancement code for speech recognition/05  Data for improvement/외국인 음성 데이터셋/Error Information data/중국어/01errorInfoDict.p'
with open(errorInfoPath, "rb") as f:
    errorInfoDict = pickle.load(f)
```

# error Tagging DataFrame 생성

```python
ErrorDict = {
    'file name' : [],
    'file length' : [],
    'number of errors' : [],
    'time of error' : [],
}
for key in tqdm_notebook(list(errorInfoDict)):
    ErrorDict['file name'].append(key)
    ErrorDict['file length'].append(len(errorInfoDict[key]))
    tempErrorTime = []
    for i in range(len(errorInfoDict[key])):
        if errorInfoDict[key][i] == 3:
            tempErrorTime.append(i)
    
    tempErrorSection = []
    for i in range(len(tempErrorTime)-1):
        check = True
        for j in range(tempErrorTime[i] + 1, tempErrorTime[i+1] - 1):
            if errorInfoDict[key][j] != 4:
                check = False
                break
        
        if check:
            tempErrorSection.append((tempErrorTime[i], tempErrorTime[i+1]))
        else:
            tempErrorSection.append((tempErrorTime[i], tempErrorTime[i]))
    
    while True:
        check = True
        for i in range(len(tempErrorSection) - 1):
            if tempErrorSection[i][1] == tempErrorSection[i + 1][0]:
                tempErrorSection[i] = (tempErrorSection[i][0], tempErrorSection[i+1][1])
                check = False
                del tempErrorSection[i+1]
                break
        if check:
            break

    for i in range(len(tempErrorSection)):
        tempSection = (tempErrorSection[i][0] * 320, (tempErrorSection[i][1] + 1) * 320)
        tempErrorSection[i] = tempSection

    ErrorDict['number of errors'].append(len(tempErrorSection))
    ErrorDict['time of error'].append(tempErrorSection)
```

```python
ErrorInfoDf = pd.DataFrame(ErrorDict)
```
